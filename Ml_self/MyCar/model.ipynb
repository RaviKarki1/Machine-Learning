```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#for making all columns visible
pd.pandas.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("car data.csv")
df.shape
```

```python
df.head()
```

Look at the types of fuelType, seller, transmissions available

```python
print(df['Fuel_Type'].unique())
print(df['Seller_Type'].unique())
print(df['Transmission'].unique())
```

visualise the relationship of these varables with the price

```python
df.columns
```

```python
cat_features = df[['Fuel_Type', 'Seller_Type', 'Transmission']]
cat_features
```

```python
for features in cat_features:
    data = df.copy()
    data.groupby(features)['Selling_Price'].median().plot.bar()
    
    plt.xlabel(features)
    plt.ylabel("selling price")
    plt.show()
```

We can see that the price is strongly dependent on these factors 
so we will consider them for our final model

Onehotencoding these values for the use in our model

```python
cat_features = pd.get_dummies(data = cat_features, )
cat_features
```

```python
labels_to_drop = ['Fuel_Type', 'Seller_Type', 'Transmission']
df.drop(['Fuel_Type', 'Seller_Type', 'Transmission'], axis = 1, inplace=True)
df.head()
```

```python
#concatinating the df with cat onehotencoded values
dataFrame = [df, cat_features]
df_1 = pd.concat(dataFrame, axis = 1)
df_1
```

```python
df_1['Year']
```

```python
df_1['cur_year'] = 2020
df_1
```

```python
df_1['years_run'] = df_1['cur_year'] - df_1['Year']
drop_cols = ['cur_year', 'Year']
df_1.drop(drop_cols, axis = 1, inplace = True)
df_1.head()
```

```python
print(df['Owner'].unique())
df_1.drop('Car_Name', axis = 1)
```

```python
df_1 = pd.get_dummies(data = df_1,)
df_1.head()
```

```python
X = df_1.drop('Selling_Price', axis = 1)
y = df_1['Selling_Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_text, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
prediction = regressor.predict(X_text)
prediction
```

```python
sns.distplot(y_test - prediction)
```

```python
plt.scatter(y_test, prediction)
```
