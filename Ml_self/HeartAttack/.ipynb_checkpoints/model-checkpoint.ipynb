```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

missing_values = ["?"]
df = pd.read_csv("data.csv", na_values = missing_values)
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.rename(columns = {"num       ":'num'})
```

```python
df.columns
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop('ca', axis = 1, inplace = True)
df.drop('thal', axis = 1, inplace = True)
```

```python
df.head()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
chol_median = df['chol'].median()
df['chol'].fillna(chol_median, inplace = True)

fbs_median = df['fbs'].median()
df['fbs'].fillna(fbs_median, inplace = True)

fbs_median = df['slope'].median()
df['slope'].fillna(fbs_median, inplace = True)

```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.corr()
```

```python
sns.pairplot(df)
```

```python
corrmat = df.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20, 20))

g = sns.heatmap(df[top_corr_features].corr(), annot = True, cmap = "RdYlGn")
```

```python
X = df.drop('num', axis = 1)
y = df.iloc[:, -1]
```

```python
for feature in df:
    data = df.copy()
    data.groupby(feature)['num'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('num')
    plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33)
```
