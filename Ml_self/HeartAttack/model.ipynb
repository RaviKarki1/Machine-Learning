```python
chol_median = df['chol'].median()
df['chol'].fillna(chol_median, inplace = True)

fbs_median = df['fbs'].median()
df['fbs'].fillna(fbs_median, inplace = True)

fbs_median = df['slope'].median()
df['slope'].fillna(fbs_median, inplace = True)

```

```python
corrmat = df.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(20, 20))

g = sns.heatmap(df[top_corr_features].corr(), annot = True, cmap = "YlGnBu")
```

```python
X = df.drop('num', axis = 1)
y = df.iloc[:, -1]
```

```python
for feature in df:
    data = df.copy()
    data.groupby(feature)['num'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('num')
    plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33)
```

```python
X_train.head()
```

```python
from sklearn import svm

classifier = svm.SVC(kernel='linear', gamma = 'auto', C = 2)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
import pickle
file = open('heartAttackModel.pkl', 'wb')
pickle.dump(classifier, file)
```

```python
open_File = open('heartAttack.pkl', 'rb')
model = pickle.load(open_File)
model.predict(X_test)
```
